# glove_vector

```python
from mmsdk import mmdatasdk as md
import pandas as pd

# 데이터셋 경로
dataset_path = '/content/CMU-MultimodalSDK/mmsdk/mmdatasdk/dataset/standard_datasets/CMU_MOSI'

# 데이터셋 로드
dataset = md.mmdataset(dataset_path)

# 사용 가능한 모든 키 출력하여 확인
#print("Available keys in dataset:", dataset.computational_sequences.keys())

# 데이터 키(시간)와 특성 확인 (올바른 키 사용)
correct_key = 'glove_vectors'  # 예시로 dict_keys 이름 넣기
for key, value in dataset.computational_sequences[correct_key].data.items():
    print(f"Time: {key}, Features: {value}")

# 메타데이터 출력
metadata = dataset.computational_sequences[correct_key].metadata
print("Metadata:", metadata)
```

```python
# 데이터프레임으로 변환
def dataset_to_dataframe(dataset):
    data_dict = {}
    correct_key = 'glove_vectors'  # 예시로 dict_keys 이름 넣기
    for key, value in dataset.computational_sequences[correct_key].data.items():
        data_dict[key] = {
            'intervals': value['intervals'],
            'features': value['features']
        }
    return pd.DataFrame.from_dict(data_dict, orient='index')

df = dataset_to_dataframe(dataset)
```

```python
# Let's take 'COVAREP' as an example
covarep_data = cmumosi_dataset['glove_vectors']
```

```python
# 'COVAREP' 데이터셋의 키를 확인합니다.
print("Keys in 'glove_vectors' dataset:", covarep_data.keys())
```

```python
# 'glove_vectors' 데이터셋의 키를 확인합니다.
keys = covarep_data.keys()

# 각 키(동영상 ID)에 대한 데이터 길이(또는 개수)를 확인합니다.
for key in keys:
    data_length = len(covarep_data[key]['features'])
    print(f"Video ID: {key}, Data Length: {data_length}")
```
